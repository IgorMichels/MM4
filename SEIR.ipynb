## Modelo SEIR 

$$\frac{dS}{dT}=B-\beta SI-\mu S$$

$$\frac{dE}{dT}=\beta SI-(\epsilon +\mu )E$$

$$\frac{dI}{dT}=\varepsilon E-(\gamma +\mu )I$$

$$\frac {dR}{dT}=\gamma I-\mu R$$

```python
def model(t, y, params):
    S, E, I, R = y
    B, beta, mu, eps, gamma = params
    return [
        B - beta * S * I - mu * S,  # dS/dt
        beta * S * I - (eps + mu) * E,  # dE/dt
        eps * E - (gamma + mu) * I,  # dI/dt
        gamma * I - mu * R,  # dR/dt
    ]  
```
