# Rascunho do modelo

Nosso modelo pode ser expresso como

$$
\begin{align}
    \dfrac{dS}{dt} & = -\dfrac{\beta(t)}{N}SI, \\
    \dfrac{dE}{dt} & = \dfrac{\beta(t)}{N}SI - \alpha(t) E, \\
    \dfrac{d\beta}{dt} & = -\dfrac{\rho_k\left(t - t_k\right)}{t^2}, \\
    \dfrac{dI}{dt} & = \alpha(t) E - \dfrac{1}{T_I}I, \\
    \dfrac{dR}{dt} & = \dfrac{1 - f(t)}{T_I}I, \\
    \dfrac{dD}{dt} & = \dfrac{f(t)}{T_I}I.
\end{align}
$$

```python
def model1(t, y, params):
    S, E, I, R, D = y
    tp, alpha, rho, ft, Ti = params
    '''
        t > 0
        S, E, I, R e D so os compartimentos usuais.
        tp  a partio do tempo que estamos fazendo (dia inicial de cada perodo)
        alpha  o vetor de alphas para cada intervalo em tp (so constantes)
        rho  o vetor dos rho_k para cada intervalo tp (usados na funo de beta)
            uma ideia  colocar beta(t_0) em rho[-1]
        f  o vetor de f(t) para cada intervalo pt (so constantes tambm)
        Ti  a taxa de mortalidade
    '''
    
    print("oi", t)
    
    N = S + E + I + R + D
    beta = 1 - rho*tp
    return [
        - beta / N * S * I,            # dS/dt
        beta / N * S * I - alpha * E,  # dE/dt
        alpha * E - I,                 # dI/dt
        (1 - ft) * I,                  # dR/dt
        ft * I                         # dD/dt
#         alpha * E - 1 / Ti * I,        # dI/dt
#         (1 - tf) / Ti * I,             # dR/dt
#         ft / Ti * I                    # dD/dt
    ]

def model(t, y, params):
    S, E, I, R = y
    B, beta, mu, eps, gamma = params
    return [
        B - beta * S * I - mu * S,  # dS/dt
        beta * S * I - (eps + mu) * E,  # dE/dt
        eps * E - (gamma + mu) * I,  # dI/dt
        gamma * I - mu * R,  # dR/dt
    ]
```

```python
tp = 0
alpha = 0.0000000007
rho   = 0.0000000001
# rho.append(0.2)
f     = 0.0000000005
Ti    = 0.0000000005
```

```python
T = ode_solver()
T.function = model1
inits = [.99, 0, 1e-6, 0, 0]
tspan = [0, 280]
T.ode_solve(tspan, inits, num_points = 280, params = [tp, alpha, rho, f, Ti])
model1 = [(k[0], k[1][3]) for k in T.solution]

T.plot_solution()
```

```python
def achaBt(tp, rho):
    beta = [rho[-1]]
    for tempo in range(len(tp)):
        if tp[tempo] != 0:
            nbeta = beta[-1]*(1 - rho[tempo]*(tp[tempo] - tp[tempo - 1])/tp[tempo])
            beta.append(nbeta)
    
    return beta
```

```python
def model2(t, y, params):
    S, E, I, R, D = y
    tp, alpha, rho, f, Ti = params
    '''
        t > 0
        S, E, I, R e D so os compartimentos usuais.
        tp  a partio do tempo que estamos fazendo (dia inicial de cada perodo)
        alpha  o vetor de alphas para cada intervalo em tp (so constantes)
        rho  o vetor dos rho_k para cada intervalo tp (usados na funo de beta)
            uma ideia  colocar beta(t_0) em rho[-1]
        f  o vetor de f(t) para cada intervalo pt (so constantes tambm)
        Ti  a taxa de mortalidade
    '''
    # criar um vetor de betas para os tempos finais de cada tp
    beta = achaBt(tp, rho)
    
    indice = 0
    while t <= tp[indice]:
        # s para saber em qual tempo estamos
        indice += 1
        
    beta = beta[indice]*(1 - rho[indice]*(t - tp[indice])/t)
    alpha = alpha[indice]
    ft = f[indice]
    
    return [
        - beta / N * S * I,  # dS/dt
        beta / N * S * I - alpha * E,  # dE/dt
        alpha * E - 1 / Ti * I,  # dI/dt
#         (1 - tf) / Ti * I,  # dR/dt
#         ft / Ti * I # dD/dt
        (1 - ft) * I,  # dR/dt
        ft * I # dD/dt
    ]
```
